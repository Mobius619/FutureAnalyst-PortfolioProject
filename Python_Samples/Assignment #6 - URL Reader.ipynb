```python
# -*- coding: utf-8 -*-
"""
Created on Sun Feb 18 17:46:46 2024

@author: Ramir Aguilo
"""

# Assignment 6: "URL Reader"

# Ramir Aguilo

import urllib
import sys


# For Error Handling
try:
    url_str = input('Please enter a text resource URL like http://textfiles.com/adventure/221baker.txt \n').strip()
    print(url_str)
    txt = urllib.request.urlopen(url_str)
except:
    print('URL not valid:', url_str)
    sys.exit()
    
# Initialize Counter
size = 0

 

# Assign values to 'Named Constants.'
CHUNK_SIZE = 512
PRINT_LIMIT = 3000

while size < PRINT_LIMIT:

    info = txt.read(CHUNK_SIZE)
    
    if len(info) < 1: break
    
    CHUNK_SIZE = min(len(info), PRINT_LIMIT - size)  # So as not to exceed limit of characters
    
    print(info[:CHUNK_SIZE].decode("utf-8", errors="ignore"))  # Decode the binary chunk to UTF-8
    
    size += CHUNK_SIZE

        
        
    print('The size of that last chunk was', CHUNK_SIZE) # For verification 
        
    
    # Open a file to 'write' to
    fhand = open('write_to_here.txt', 'w')
    
    fhand.write(info.decode("utf-8", errors="ignore"))

    
    print(f"{size} characters read so far.")
    print('\n')  # Separate chunks


print(size, 'characters copied.')


    

      
      
```
